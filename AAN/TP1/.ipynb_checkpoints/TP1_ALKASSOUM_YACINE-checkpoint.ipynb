```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

## <span style="color:brown; font-family:times; font-size:1.3em;">3.2 Load the training and testing data </span> 

```python
houses_train = pd.read_csv('train.csv')
houses_train.head()
```

```python
houses_test= pd.read_csv('test.csv')
houses_test.head()
```

```python
houses_train.dtypes
houses_train.describe()
```

## <span style="color:brown; font-family:times; font-size:1.3em;">3.3 Variables Correlation with SalePrice </span>
<span style="color:black; font-family:Georgia; font-size:1.3em;">In this section we will search for the two most correlated variables.</span>


```python
# Todo
# calculate the correlation between salePrice and all variables
# Keep the two most correlated variables to test them with the simple linear regression

x = houses_train.corr(method='spearman').get('SalePrice')
print(x.sort_values().drop("SalePrice").tail(2))
```

## <span style="color:brown; font-family:times; font-size:1.3em;">3.4 Plot SalePrice    </span> 

```python
# Todo
# plot the SalePrices as function of your 2 variables (most correlated)
# add x label and Y label, the values and the scatter
livA=[x for x in houses_train.get('GrLivArea')]
qual=[x for x in houses_train.get('OverallQual')]
salesP=[x for x in houses_train.get('SalePrice')]
plt.scatter(livA,salesP,c='red')
plt.figure()
plt.scatter(qual,salesP,c='blue')
plt.title("Distribution of Sale Price as function of  ")
plt.xlabel("Saleprice")
plt.ylabel("Sale Price (in dollar)");
```

# <span style="color:green; font-family:Georgia; font-size:1.5em;">4. Regression Lineaire Simple (Python) </span>


```python
# Todo 

# Estimation de \beta_1
def b1(x,y):

    ybarre=np.mean(y)
    xbarre=np.mean(x)
    B1=np.cov(x,y,ddof=0)[0][1]/np.var(x)
    print("B1:",B1)
    return B1
# 
B1livA=b1(livA,salesP)
B1qual=b1(qual,salesP)
```

```python
# Todo 
# Estimation de \beta_0
def b0(x,y):

    ybarre=np.mean(y)    
    xbarre=np.mean(x)
    B1=np.cov(x,y,ddof=0)[0][1]/np.var(x)
    Bo = ybarre - B1*xbarre
    print("BO:",Bo)
    return Bo
B0livA=b0(livA,salesP)
B0qual=b0(qual,salesP)
plt.scatter(livA,salesP,c='red')
plt.plot(livA,[B1livA*x+B0livA for x in livA])
plt.figure()
plt.scatter(qual,salesP,c='blue')
plt.plot(qual,[B1qual*x+B0qual for x in qual])
plt.title("Distribution of Sale Price as function of  ")
plt.xlabel("Saleprice")
plt.ylabel("Sale Price (in dollar)");
```

```python
# Calculez le coeffecient de détermination R2 
def coeffR2(y,ychapo,ybarre):
    SCE=0
    SCT=0
    for cpts,cptchapo in zip(y,ychapo):
        SCE+=pow((cptchapo-ybarre),2)
        SCT+=pow((cpts-ybarre),2)
    R2=SCE/SCT
    return R2
R2_livA=coeffR2(salesP,[B1livA*x+B0livA for x in livA],np.mean(salesP))
R2_qual=coeffR2(salesP,[B1qual*x+B0qual for x in qual],np.mean(salesP))
print("GrLivArea",R2_livA)
print("OverallQual",R2_qual)



```

```python
# prédire les valeurs sur les données de Test
houses_test.dtypes
houses_test.describe()


testlivA=[x for x in houses_test.get('GrLivArea')]
testqual=[x for x in houses_test.get('OverallQual')]
testsalesP=[x for x in houses_test.get('SalePrice')]
plt.scatter(testlivA,testsalesP,c='red')
plt.plot(testlivA,[B1livA*x+B0livA for x in testlivA])
plt.figure()
plt.scatter(testqual,testsalesP,c='blue')
plt.plot(testqual,[B1qual*x+B0qual for x in testqual])
plt.title("Distribution of Sale Price as function of  ")
plt.xlabel("Saleprice")
plt.ylabel("Sale Price (in dollar)");
```

# <span style="color:green; font-family:Georgia; font-size:1.5em;">5.  Evaluation du  modèle (RMSE)
  </span>


<span style="font-family:Georgia; font-size:1.2em;">La  racine  carrée  de l’erreur quadratique moyenne (RMSE) est une   autre  mesure  des  erreurs  de  prévision. La RMSE est définie par : $ RMSE = \sqrt{ \frac{1}{n} \sum (pred - y)^{2}} $.
</span>

<span style="font-family:Georgia; font-size:1.2em;">
Cette  mesure  pénalise  d'avantage  les  erreurs  importantes 
</span>

```python
# Calculez la RMSE de votre modèle sur les données de test
def RMSE(pred, Y, n):
    return np.sqrt((1/n)*sum((pred-Y)**2))

rmselivA=sqrt(mean_squared_error(testsalesP, [B1livA*x+B0livA for x in testlivA]))
rmsequal=sqrt(mean_squared_error(testsalesP,[B1qual*x+B0qual for x in testqual]))
print("RMSE GrLivArea on test csv :",rmselivA)
print("RMSE OverallQual on test csv:",rmsequal)
```

# <span style="color:green; font-family:Georgia; font-size:1.5em;">5.  Pour aller plus loin </span>

<span style="font-family:Georgia; font-size:1.2em;">  
La standardisation des données consiste à centrer-réduire la matrice de fonctions caractéristiques  $X$ en retranchant pour chaque vecteur $X^{j}, j \in [1, N]$ sa moyenne et en divisant par son écart-type.
</span>


<span style="font-family:Georgia; font-size:1.2em;">
La formule d’usage est alors : $ ~~~~~~~~~~~  \forall j \in [1, N] ~~~~ X^{j} = \frac{X^{j}_{std} - \mu_{j}}{ \sigma^{2}_{j}}$

</span>

```python
# Todo :
##  refaire le calcul des paramètres après un prétraitement sur les données d'apprentissage
##  1.  Centrer la variable --> soustraire son espérance à chacune de ses valeurs initiales
##  2.  Réduire la variable --> diviser toutes ses valeurs par son écart type.
## --------------------------
##    Soient  μ  l’espérance et σ l’écart-type des valeurs d’une variable aléatoire.
##    Centrer-réduire l’une de ses valeurs V revient alors à calculer : 
##          (V - μ ) / σ 
## --------------------------
##  3. re-éstimer les paramètre de modèle 
##  4. re-calculer la précision du modèle sur les données de test.

#1.
moylivA=np.mean(houses_train["GrLivArea"])
moyqual=np.mean(houses_train["OverallQual"])
moySaleP=np.mean(houses_train["SalePrice"])
ElivA=np.std(houses_train["GrLivArea"])
Equal=np.std(houses_train["OverallQual"])
EsaleP=np.std(houses_train["SalePrice"])

testmoyqual=np.mean(houses_test["OverallQual"])
testmoylivA=np.mean(houses_test["GrLivArea"])
testmoySaleP=np.mean(houses_test["SalePrice"])
testEqual=np.std(houses_test["OverallQual"])
testElivA=np.std(houses_test["GrLivArea"])
testEsaleP=np.std(houses_test["SalePrice"])


nrmlsedLivA=[(i-moylivA)/ElivA for i in houses_train["GrLivArea"]]
nrmlsedQual=[(i-moyqual)/Equal for i in houses_train["OverallQual"]]
nrmsedSaleP=[(i-moySaleP)/EsaleP for i in houses_train["SalePrice"]]

testnrmlsedQual=[(i-testmoyqual)/testEqual for i in houses_test["OverallQual"]]
testnrmlsedLivA=[(i-testmoylivA)/testElivA for i in houses_test["GrLivArea"]]
testnrmsedSaleP=[(i-testmoySaleP)/testEsaleP for i in houses_test["SalePrice"]]




print("Beta1 pour GrLivArea")
nrmlivAB1=b1(nrmlsedLivA,nrmsedSaleP)

print("Beta1 pour OverallQual")
nrmqualB1=b1(nrmlsedQual,nrmsedSaleP)
print("\nBeta0 pour GrLivArea")
nrmlivAB0=b0(nrmlsedLivA,nrmsedSaleP)
print("Beta0 pour OverallQual")
nrmqualB0=b0(nrmlsedQual,nrmsedSaleP)


nqualR2=coeffR2(nrmsedSaleP,[nrmqualB1*x+nrmqualB0 for x in nrmlsedQual],np.mean(nrmsedSaleP))
nlivAR2=coeffR2(nrmsedSaleP,[nrmlivAB1*x+nrmlivAB0 for x in nrmlsedLivA],np.mean(nrmsedSaleP))

print("\nR2 pour OverallQual",nqualR2,"R2 pour GrLivArea",nlivAR2)

#RMSE
rmsenlivA=sqrt(mean_squared_error(testnrmsedSaleP, [nrmlivAB1*x+nrmlivAB0 for x in testnrmlsedLivA]))
rmsenqual=sqrt(mean_squared_error(testnrmsedSaleP, [nrmqualB1*x+nrmqualB0 for x in testnrmlsedQual]))
print("\nRMSE : \nOverallQual:",rmsenqual," \nGrLivArea:",rmsenlivA)


```
